```python
# YOUR CODE GOES HERE 
def averageRegionPop(region, year):
    indDf = pd.read_csv("indicators.csv")
    countryDf = pd.read_csv("countries.csv")
    
    indDf.rename(columns = {"code" : "ccode"}, inplace=True)
    combinedDf = pd.merge(indDf, countryDf, on="ccode", how="inner")
    
    
    ypcrDf = combinedDf.loc[(combinedDf["region"] == region) &(combinedDf["year"] == year),["year","pop","country","region"]]
    
    regionYearGroup = ypcrDf.groupby(["region", "year"])
    averagePop = regionYearGroup.agg({"pop":["mean"]})
    sortedAvgPop = averagePop.sort_values(("pop", "mean"), ascending=False)
    
    return sortedAvgPop.iat[0,0]

averageRegionPop("Europe & Central Asia", 1999)
```

* <code>exports:</code> total exports given in <code>indicators.json</code> (in millions of dollars).
* <code>imports:</code> total imports given in <code>indicators.json</code> (in millions of dollars).
* <code>net exports:</code> exports minus imports (in millions of dollars).

Furthermore, the rows of the table should be ordered alphabetically by country name. You may use whatever method you like to create the Data Frame (however, the problem should be much easier now that you've learned pandas!)
```python
# YOUR CODE GOES HERE 
def dodToLod(dod):
    '''
    Given a dod (hierarchical) representation of the indicators data, 
    converts the data to a lod (tabular) representation. 
    
    Parameters: dod representation of data
    Return Value: lod representation of data
    '''
    lod = []      
    for code in dod:                      # for each country code in data set (top level branching)
        for year in dod[code]:            # for year for code subtree (second level branching)
            rowdict = {}                  # consturct lod row to go into lod structure 
            rowdict["code"] = code    
            rowdict["year"] = year
            for stat in dod[code][year]:                # add all statss in current year subtree to row 
                rowdict[stat] = dod[code][year][stat]
            lod.append(rowdict)                         # add row to lod 
    return lod
```
